```python
import matplotlib
%matplotlib inline

import os
import requests

import json
from pprint import pprint

from bs4 import BeautifulSoup
import pandas as pd

import glob
```

# CSV Consolidation

```python
filenames = glob.glob("*.csv")
```

```python
list_of_dfs = [pd.read_csv(filename) for filename in filenames]
```

```python
for dataframe, filename in zip(list_of_dfs, filenames):
    dataframe['Filename'] = filename
```

```python
df = pd.concat(list_of_dfs, ignore_index=True)
```

```python
df
```

```python
df['Filename'] = df['Filename'].str.replace('(.csv)', '', regex = True)
```

```python
df
```

```python
df.to_csv("All.csv", index=False)
```

# Cleaning

```python
df_clean = df
df_clean.dropna(subset=['Location'], inplace=True)
```

```python
df_clean
```

```python
df_clean.to_csv("Clean.csv", index=False)
```
