```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
#import re
#from collections import Counter
#import numpy as np
import nltk
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42 #makes text editable in illustrator

%matplotlib inline
```

```python
df = pd.read_csv("turnout.csv", encoding='latin-1')
```

```python
df.head()
```

```python
df = df.rename(index=str, columns={df.columns[0]: "VEP Year"})
df = df.rename(index=str, columns={df.columns[1]: "VEP Turnout"})
df = df.rename(index=str, columns={df.columns[2]: "Midterm Year"})
df = df.rename(index=str, columns={df.columns[3]: "Midterm Turnout"})
```

```python
df.head()
```

```python
df.dtypes
```

```python
ax = df.sort_values(by='VEP Year', ascending=True).plot(x='VEP Year', 
           y='VEP Turnout', 
           figsize=(10, 6),
           title="VEP Year",
           color="#5EAE9B")
plt.tight_layout()
plt.savefig("VEP.pdf")
```

```python
ax = df.sort_values(by='Midterm Year', ascending=True).plot(x='Midterm Year', 
           y='Midterm Turnout', 
           figsize=(10, 6),
           title="Midterm Year",
           color="#5EAE9B")
plt.tight_layout()
plt.savefig("Midterm.pdf")
```
